```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,40))

for i , val in enumerate(df.keys()):
    plt.ylabel(val)
    plt.subplot(12 , 4 ,i+1)
    sns.distplot(df[val] , bins=25)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr())
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[: , :-1]
Y = df.iloc[: , -1]

x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.2)
print(x_train.shape , x_test.shape , y_train.shape , y_test.shape)
```

# using rbf (radial basis function) 

```python
from sklearn.svm import SVC

sc = SVC(kernel='rbf')
clf = sc.fit(x_train , y_train)
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , precision_score, recall_score , precision_recall_curve

cnf_mat = confusion_matrix(y_pred , y_test)
print("Accuracy : " , accuracy_score(y_pred , y_test))
print("Pression_score : " , precision_score(y_pred , y_test))
print("Precision_recall : " , recall_score(y_pred , y_test))
```

```python
precision_recall_curve(y_pred , y_test)
```

```python
sns.heatmap(cnf_mat)
```

# Using Linear

```python
sc = SVC(kernel='linear')
clf = sc.fit(x_train , y_train)
y_pred = clf.predict(x_test)
y_pred
```

```python
cnf_mat = confusion_matrix(y_pred , y_test)
print("Accuracy : " , accuracy_score(y_pred , y_test))
print("Pression_score : " , precision_score(y_pred , y_test))
print("Precision_recall : " , recall_score(y_pred , y_test))
```

```python
sns.heatmap(cnf_mat)
```
