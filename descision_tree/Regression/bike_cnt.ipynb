# splitting

```python
from sklearn.model_selection import train_test_split

x_day = df_day.iloc[:,:-1].values
x_hour = df_hour.iloc[:,:-1].values
y_day = df_day.iloc[: , -1].values
y_hour = df_hour.iloc[: , -1].values

xd_train , xd_test , yd_train , yd_test = train_test_split(x_day , y_day , test_size=0.05)
xh_train , xh_test , yh_train , yh_test = train_test_split(x_hour , y_hour, test_size=0.05)
print(yd_train.shape , yd_test.shape )
```

# Model Creation

```python
from sklearn.tree import DecisionTreeRegressor

clf_hr = DecisionTreeRegressor(criterion='friedman_mse')
clf_dy = DecisionTreeRegressor(criterion='friedman_mse')
clf_dy = clf_dy.fit(xd_train , yd_train)
clf_hr = clf_hr.fit(xh_train , yh_train)
```

# Prediction

```python
yh_pred = clf_hr.predict(xh_test)
print(xd_test.shape  , xd_train.shape)
# print("Prediction of day -> ", yd_pred)
print("prediction of hr -> ",yh_pred)
```

```python
yd_pred = clf_dy.predict(xd_test)
print(yd_pred , yd_test)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(yd_test , yd_pred))
```

```python
sns.heatmap(confusion_matrix(yh_test , yh_pred))
```
